Fundamental Theorem of Algebra

See "What is Mathematics?" page 271.
The proof uses polynomial mappings of circles in the complex plane. Let's visualize some.

The thought goes like this: we are trying to prove that all polynomials have at least one root. 
Given a polynomial $f$, we assume $f(0) \neq 0$ 
(otherwise the constant term is zero, we have a root and we are done) 
then evaluate $f$ along a circle $z$ in the complex plane with radius $t$. 
For small values of $t$, $f(z)$ will make a small closed loop away from the origin. 
For large values of $t$, $f(z)$ will make a larger closed loop that contains the origin. 
Since everything we are dealing with here is continuous, 
we can conclude that for some circle $z$ in the complex plane, 
$f(z)$ crosses the origin, and that's our root. 
(Of course, this is not a proof, I'm just doing this to help me visualize what's going on.)

```python
import matplotlib.pyplot as plt
import numpy as np

# data setup
line = np.arange(0,2*np.pi,.1)
z = 0.1*(np.cos(line) + 1j*np.sin(line))

#"pz" = p(z) where p is given polynomial
p = [1, 3, -2]
pz = np.polyval(p, z)

```

```python
# real case just for fun
line = np.arange(-5,5,0.1)
px = np.polyval(p, line)
plt.plot(line, px)
plt.show()
```

```python
# plot for some small z
plt.plot(pz.real, pz.imag)
plt.plot(z.real, z.imag)
plt.show()
```

```python
# show circles getting bigger
p = [1, 7+3j, 3, -2, 4]
r = [0.01, 0.1, 0.3, 0.54, 0.7, 1, 1.2]

for t in r:
    z = t*(np.cos(line) + 1j*np.sin(line))
    pz = np.polyval(p, z)
    plt.plot(pz.real, pz.imag)
    plt.plot(z.real, z.imag)

plt.show()



```

Yes, eventually we cross the origin.
